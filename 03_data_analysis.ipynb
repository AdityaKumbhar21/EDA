# **Numerical and Categorical Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/adityakumbhar/Developer/Datasets/Churn_Modelling.csv')
df.head()
```

```python
new_df = df[['Geography', 'Gender', 'Exited']]
new_df.head()
```

## **Univariate Analysis**

```python
plt.figure()
sns.countplot(data=new_df, x='Geography', hue="Exited")
```

```python
plt.figure()
sns.countplot(data=new_df, x='Gender', hue="Exited")
```

## **Bivariate Analysis**

```python
new_target_1 = df.loc[df['Exited'] == 1]
new_target_1.head()
```

```python
plt.figure()
sns.histplot(x='Gender', hue='Geography', data=new_df,stat='count', multiple='dodge')
```

```python
plt.figure()
sns.histplot(x='Gender', hue='Geography', data=new_target_1,stat='count', multiple='dodge')
```

## **Numerical Analysis**

```python
df_numeric = df.select_dtypes(include='number')
df_numeric.corr()
```

```python
plt.figure(figsize=(20,8))
df_numeric.corr()['Exited'].sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(data=df_numeric.corr(), cmap="Paired")
```

```python
Tot = sns.kdeplot(df_numeric.Age[(df_numeric['Exited'] == 0)], color='Red', shade=True)
Tot = sns.kdeplot(df_numeric.Age[(df_numeric['Exited'] == 1)], color='Blue', shade=True)
Tot.legend(['No Churn', 'Churn'])
Tot.set_xlabel("Density")
Tot.set_ylabel("Age")
Tot.set_title("Churn Rate by age")
```
