```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df = pd.read_csv('/Users/adityakumbhar/Developer/Datasets/Churn_Modelling.csv')
df.head()
```

## **Feature Binning**

```python
df.info()
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'], axis=1, inplace=True)
df.info()
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```

```python
labels = ['0-10', '11-20', '21-30','31-40','41-50','51-60','60+']
bins = [0,10,20,30 ,40,50, 60, 100]

df['Age_bins'] = pd.cut(df.Age, bins, labels=labels, include_lowest=True)
df.head()
```

```python
df['Age_bins'].value_counts().sort_index()
```

## **Feature Encoding**

### **Label Encoder**.  

Label Encoding is done for y variables only not recommended for the x variables.

```python
le = preprocessing.LabelEncoder()

df['gender_label'] = le.fit_transform(df['Gender'].values)

df.head()
```

```python
df['Gender'].value_counts()
```

```python
df['gender_label'].value_counts()
```

### **One Hot Encoding**

```python
one_hot = pd.get_dummies(df['Geography'])
```

```python
one_hot
```

```python
df_dummies = pd.get_dummies(df)
df_dummies.head()
```

### **Dummy Encoding**

In this type of encoding we drop any one of the category column which is not necessarily required.

```python
df_dummies = pd.get_dummies(df, drop_first=True)
df_dummies.head()
```
